```python
# Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
# function.

# Defination: when we use yield keyword, that function became generator function automatically. Thoughtout, genreator function dont save
# data in the memory and give data one by one.
def generator_test(arg1, arg2) :
    something = "shifali"
    while arg1 <= arg2 :
        yield something+'-'+str(arg2)
        arg2 = arg2-1

generator_test(1,2)

for i in generator_test(0,10):
    print(i)
```

```python
#Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
# first 20 prime numbers.

def prime_number(num, b) :
    while num < b :
        flag = False
        if num == 1:
            yield num, "is not a prime number..."
        if num > 1 :
            for i in range(2, num) :
                if (num % i) == 0 :
                    flag = True
                break
            if flag:
                yield num, "is not a prime number..."
            else:
                yield num, "is a prime number..."
        num = num+1
for i in prime_number(1, 1001) :
    print(i)
```

```python
# Q6. Write a python program to print the first 10 Fibonacci numbers using a while loop.

count = 0
a,b = 0,1
while count < 10 :
    print(a)
    c = a+b
    a = b
    b = c
    count = count + 1
```
